```python
indicators = [
    "IT.NET.USER.ZS",
    "NY.GDP.PCAP.CD",
    "SP.POP.TOTL"
]

filtered = df[df["Indicator Code"].isin(indicators)]

filtered.to_csv("WDI_Selected.csv", index=False)
```

```python
df.info()
```

```python
df1 = pd.read_csv("WDI_Selected.csv")

df1.shape

df1.head()
```

```python
df1.info()
```

```python
country = pd.read_csv("WDICountry.csv")

country_small = country[
    ["Country Code", "Region", "Income Group"]
]

df2 = df1.merge(
    country_small,
    on="Country Code",
    how="left"
)
```

```python
df2.head()
```

```python
df3 = df2.melt( id_vars=[ 
    "Country Name", "Country Code", "Indicator Name", "Indicator Code", "Region", "Income Group" ], 
    var_name="Year", value_name="Value" )
```

```python
df3.head()
```

```python
df3 = df3[df3["Region"].notna()]
```

```python
df3.head()
```

```python
df3.to_csv("WDI_Cleaned.csv", index=False)
```
